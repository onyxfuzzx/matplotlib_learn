```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x, y, '-g')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("First Matplotlib")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y , "r")

plt.subplot(1, 2, 2)
plt.plot(y, x, "y")
```
